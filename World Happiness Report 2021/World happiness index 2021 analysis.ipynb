# Visualization

```python
# Country vs Ladder Name
```

```python
plt.figure(figsize=(30,10))
plt.scatter(x='Country name', y='Ladder score',data=Report, color='red',)
plt.xlabel("Country name")
plt.ylabel("Ladder score")
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(30,10))
sns.barplot(x='Country name',y='Ladder score',data=Report)
plt.xticks(rotation=90)
```

```python
#Country vs Logged GDP per capita
```

```python
plt.figure(figsize=(30,10))
plt.scatter(x='Country name', y='Logged GDP per capita',data=Report, color='red')
plt.xlabel("Country name")
plt.ylabel("Logged GDP per capita")
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(30,10))
sns.barplot(x='Country name',y='Logged GDP per capita',data=Report)
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(6,5))
sns.regplot(x='Ladder score',y='Logged GDP per capita',data=Report)
plt.ylim(0,)
```
